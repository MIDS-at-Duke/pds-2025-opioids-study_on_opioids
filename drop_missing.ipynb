```python
import pandas as pd
import numpy as np

pd.set_option("mode.copy_on_write", True)
```

```python
df = pd.read_csv("data/clean/final_merged.csv")

threshold = 20000
df_dropped = df[df["Population"] >= threshold]
```

```python
df_dropped.sample()
```

```python
df_dropped = df_dropped.groupby(
    ["Buyer State", "Buyer County", "Year"], as_index=False
)["Total_MME"].sum()
```

```python
df_dropped.to_csv("data/clean/threshold.csv", index=False)
```
