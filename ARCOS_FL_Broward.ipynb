```python
# Load libraries
import pandas as pd
import numpy as np

pd.set_option("mode.copy_on_write", True)
```

```python
filepath = "/Users/brucechen/Documents/IDS720 PDS/arcos_all.zip"

# Read only first 50 rows
sample = pd.read_csv(filepath, compression="zip", nrows=50, delimiter="\t")

# Check data
sample.columns
```

```python
selected_cols = [
    "BUYER_NAME",
    "BUYER_STATE",
    "BUYER_COUNTY",
    "DRUG_NAME",
    "TRANSACTION_DATE",
    "MME_Conversion_Factor",
    "CALC_BASE_WT_IN_GM",
    "MME",
    "DOSAGE_UNIT",
    "Dosage_Strength",
    "Combined_Labeler_Name",
    "Reporter_family",
]
```

```python
chunks = []

# process the data in chunks
for i, chunk in enumerate(
    pd.read_csv(
        filepath,
        chunksize=800_000,
        compression="zip",
        usecols=selected_cols,
        delimiter="\t",
        low_memory=True,
    )
):
    # filter for selected states
    broward = chunk[
        (chunk["BUYER_STATE"] == "FL") & (chunk["BUYER_COUNTY"] == "BROWARD")
    ]

    # extract year from date
    broward = broward["YEAR"] = pd.to_datetime(broward["TRANSACTION_DATE"]).dt.year

    chunks.append(broward)
    print(f"Processed chunk {i+1}")

df = pd.concat(chunks, ignore_index=True)
```

```python
df.to_csv("FL_Broward_full_data.csv", index=False)
```

```python
# export to Parquet file in the same directory as the notebook
import os

# Example new file name
new_filename = "arcos_processed.parquet"

# Get current notebook directory
notebook_dir = os.getcwd()

# Create full path with new filename
parquet_path = os.path.join(notebook_dir, new_filename)

# Save DataFrame to the new Parquet file
df.to_parquet(parquet_path, index=False)

print(f"DataFrame saved to {parquet_path}")
```
